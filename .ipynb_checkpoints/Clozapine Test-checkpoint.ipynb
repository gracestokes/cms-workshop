```python
import os
import numpy as np
import matplotlib.pyplot as plt

clozapine = os.path.join('DOPC_clozapine_only.csv')
NDclozapine = os.path.join('avg5_DOPC_NDMC_last_three_columns.csv')

molecules = [clozapine,NDclozapine]
colors = ['r','b']
n = 0

for molecule in molecules:
    contents = np.genfromtxt(fname=molecule, delimiter=',', dtype = 'unicode')
    if molecule == NDclozapine:
        contents = contents[:,3:]
    print(contents)
    headers = contents[0,:]
    data = contents[1:,:].astype(np.float)
    X = data[:,0]
    Y = data[:,1]
    error = data[:,2]






    plt.plot(X,Y,'.',label=headers[1], color = colors[n]) #data points
    plt.plot(X,Y+error,'_', color = colors[n]), plt.plot(X,Y-error,'_', color = colors[n]) #slightly janky way to plot error
    n += 1

```
