## Plotting data using matplotlib

```python
import os
import numpy
distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)

#this initializes a new figure
# if you don't use plt, this is what you would type
#matplotlib.pyplot.figure
#matplotlib.pyplot.plot(data[:,1])

import matplotlib.pyplot as plt
#why is there a pyplot? this is kind of like numpy.whatever the function you want to use is called
#what does as plt mean? this allows us to define a shortcut for the library name
```

```python
plt.figure()
plt.plot(data[:,1])
```

## Now I am on Plotting with x and y

```python
# the plot function creates a line plot. there is only one argument in the function. y is the variable and x is just a count (0 through ...). What if we want two variables: x and y?
plt.figure()
plt.plot(data[:,0],data[:,1])

# for example, this is a slicing syntax that prints every other row (increment of 2) to the 10th row 
print(data[0:10:2,:])
#this command prints every 100th point in the first column
print(data[0::100,0])
```

```python
# this is a new slicing syntax that plots every 100th data point (frame)
plt.figure()
plt.plot(data[0::100,0],data[0::100,1])
```

```python
#changing plot appearance
plt.figure()
plt.plot(data[0::100,0],data[0::100,1],'--')
```

## How to make multiple plots into one figure with subplot

```python
# for example, I will make four plots which are all part of the same figure.
# this command makes 4 graphs all in one column
fig,ax = plt.subplots(len(headers)-1,1)
ax.shape
```

```python
fig,ax2 = plt.subplots(2,2)
ax2.shape
```

```python
fig, ax = plt.subplots(len(headers)-1,1,figsize=(10,10),sharey=True)
#fig.set_figheight(10)
for col in range(1,len(headers)):
    sample = headers[col]
    ax[col-1].plot(data[0::100,0], data[0::100,col], label=sample)
    ax[col-1].set_xlabel('Simulation Frame')
    ax[col-1].set_ylabel('Distance (angstrom)')
    ax[col-1].legend()
```
